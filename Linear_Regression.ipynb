```python
import numpy as pd
import matplotlib.pyplot as plt
import pandas as pd
from sklearn import linear_model

dataset = pd.read_csv("/content/Home_price.csv")
```

```python
dataset
print(dataset.columns)
dataset.columns = dataset.columns.str.strip()
```

```python
%matplotlib inline
plt.xlabel('area')
plt.ylabel('price')
plt.scatter(dataset.area,dataset.price,color='red',marker='+')
```

```python
new_df=dataset.drop('price',axis='columns')
new_df
```

```python
model=linear_model.LinearRegression()
model.fit(new_df,dataset.price)
```

```python
model.predict([[5000]])
```

```python
model.coef_
```

```python
model_intercept=model.intercept_
model_intercept
```

```python
Capita_df=pd.read_csv("/content/canada_per_capita_income.csv")
Capita_df.columns
```

```python
New_capita_df=Capita_df['year']
```

```python
New_capita_df
```

```python
model=linear_model.LinearRegression()
Capita_df.columns = Capita_df.columns.str.replace(' ', '_').str.replace('(', '').str.replace(')', '')
X = New_capita_df.values.reshape(-1, 1)  # Reshaping to 2D (n_samples, 1)

# y remains 1D
y = Capita_df['per_capita_income_US$']

model.fit(X, y)
```
